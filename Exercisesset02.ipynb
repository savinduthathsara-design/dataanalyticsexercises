### 01. Import pandas and load dataset 

```python
import numpy as np
import pandas as pd

np.set_printoptions(suppress=True)
pd.set_option('display.float_format', '{:.2f}'.format) 
```

```python
df = pd.read_csv("exercises2_datafiles/loans.csv")

# prining the head of the data 
df.head
```

### 02. remove unrealistically large loans 

```python
df = df[df["Current Loan Amount"] < 99999999]

df.head
```

### 03. loans that have highest credit balance

```python
df.nlargest(5, "Current Loan Amount")
```

### 04. Loans with smallest annual income 

```python
df.nsmallest(5, "Annual Income")
```

### 05. Min, max, mean and quantiles (25%, 50%, 75%) 

```python
df["Current Loan Amount"].describe()
```

```python
print('Min:  ', df['Current Loan Amount'].min())
print('Max:  ', df['Current Loan Amount'].max())
print('Mean: ', df['Current Loan Amount'].mean())
print('25%:  ', df['Current Loan Amount'].quantile(0.25))
print('50%:  ', df['Current Loan Amount'].quantile(0.50))
print('75%:  ', df['Current Loan Amount'].quantile(0.75))
```

### 6. Create a new field: Actual Annual Income

```python
df["Actual Annual Income"] = df["Annual Income"] - (12 * df["Monthly Debt"])

df[["Annual Income", "Monthly Debt", "Actual Annual Income"]].head(10)
```

### 07. Long term loans quantity

```python
count = df["Term"].value_counts()
print(count)
print(f"Number of long term loans : {count.get("Long Term")}")
```

### 08. 3 most common loan purposes 

```python
# replace the "other" with "Other" 
df["Purpose"] = df["Purpose"].replace("other", "Other")

# get all purposes 
purpose_count = df["Purpose"].value_counts()
print(purpose_count)

# top 3 most common purposes
print(f"\n {purpose_count.head(3)}")
```

### Extra task: Distribution of Number of Credit Problems

```python
credit_problems = df["Number of Credit Problems"].value_counts()
print(credit_problems)
```

```python
# Findings:
# The majority of borrowers have 0 credit problems.
# Very few borrowers have 2 or more problems. This means that most customers have clean credit histories.
```

### 09. Unique Home Ownership options

```python
unique_values = df["Home Ownership"].unique()
print(unique_values)
```

### Extra task: How many loaners have more than 1 bankruptcy

```python
bankruptcy = df[df["Bankruptcies"] > 1]
print(f"loaners have more than 1 bankruptcy : {len(bankruptcy)}")
```

### Extra task: How many Short Term loans are for Home Improvements?

```python
# filter short term loans 
short_term = df[df['Term'] == 'Short Term']

# short term loan purposes
short_term_purposes = short_term["Purpose"].value_counts()
print(short_term_purposes)

# home improvement purpose
home_improvement_count = short_term_purposes.get('Home Improvements')
print(f"\nShort term loans for home improvements: {home_improvement_count}")
```

### Extra task: Correlation Analysis

```python
# Option A: Annual Income vs Number of Open Accounts ---
corr_a = df['Annual Income'].corr(df['Number of Open Accounts'])
print(f'Correlation (Annual Income vs Number of Open Accounts): {corr_a:.4f}')
```

**Annual Income vs Number of Open Accounts:** A weak correlation suggests that having more open accounts does not strongly correlate with higher or lower income.

```python
# Option B: Number of Credit Problems vs Bankruptcies ---
corr_b = df['Number of Credit Problems'].corr(df['Bankruptcies'])
print(f'Correlation (Number of Credit Problems vs Bankruptcies): {corr_b:.4f}')
```

**Number of Credit Problems vs Bankruptcies:** this correlation would confirm that borrowers with more credit problems also tend to have more bankruptcies. This is true because bankruptcies are one of the most severe forms of credit problems.
