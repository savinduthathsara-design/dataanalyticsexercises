```python
# load the dataset
df = pd.read_csv("exercises2_datafiles/furniture.csv")

df.head()
```

### 01. Duplicate values

```python
duplicates = df.duplicated().sum()
print(f"Number of duplicate rows: {duplicates}")
df = df.drop_duplicates()

df.shape
```

### 02. NaN values

```python
Nan_values = df.isna().sum()
print("Number of NaN values in each column:")
print(Nan_values)
```

### 03. Impute NaN prices using profit multiplier (price / cost)

```python
# I used AI to figure out how to approch to this problem and how to filter NaN values
# and input the values we create 

# profit multiplier (price / cost) per category
category_multiplier = (df["price"] / df["cost"])
category_multiplier = category_multiplier.groupby(df['category']).mean()

# Fill NaN prices using formula (cost * category multiplier)
formula = df["cost"] * df["category"].map(category_multiplier)
df['price'] = df['price'].fillna(formula)

print("profit multiplier per category:")
print(category_multiplier.round(3))
print(f"\nNaN values in price after multiplier imputation: {df["price"].isnull().sum()}")
```

### 04. Create delivery category column

```python
# create the function 
def categorize_delivery(days):
    if days < 4:
        return "quick"
    elif days <= 7:
        return "normal"
    else:
        return "slow"
    
df["delivery_category"] = df["delivery_days"].apply(categorize_delivery)

df.head()
```

```python
# remove delivery days column
df = df.drop(columns = ["delivery_days"])
df.head()
```

### 05. Calculate actual discount amount

```python
# create discount formula 
discount_formula = df["price"] * df["discount_percentage"] / 100
df["discount_amount"] = discount_formula.round(2)

df.head()
```

```python
# remove discount_percentage column
df = df.drop(columns = ["discount_percentage"])
df.head()
```

### Load and perpare salaries_india dataset using previous exercise 

```python
# Load salaries_india dataset
df_salaries = pd.read_csv("exercises2_datafiles/salaries_india.csv")

# Drop unnecessary columns
df_salaries = df_salaries.drop(columns=["Row ID", "Job Title"])

# Remove outliers (salaries > 4,000,000)
df_salaries = df_salaries[df_salaries["Salary"] <= 4000000]

# Encode categorical variables
df_salaries["Employment Status"] = LabelEncoder().fit_transform(df_salaries["Employment Status"])

# Encode Rating (ordinal)
rating_order = {"Bad": 1, "Poor": 2, "Average": 3, "Good": 4, "Excellent": 5}
df_salaries["Rating"] = df_salaries["Rating"].map(rating_order)

# Encode Job Roles and Location (nominal)
df_salaries = pd.get_dummies(df_salaries, columns=["Job Roles", "Location"])

print(f"Prepared dataset shape: {df_salaries.shape}")
df_salaries.head()
```

### Compare Phik vs Correlation with Salary

```python
corr_with_salary = df_salaries.corr(numeric_only=True)['Salary'].drop('Salary').sort_values(ascending=False)
phik_df = df_salaries.phik_matrix()
phik_with_salary = phik_df['Salary'].drop('Salary').sort_values(ascending=False)

# Create comparison DataFrame
comparison = pd.DataFrame({
    "Correlation": corr_with_salary,
    "Phik": phik_with_salary
})

print(comparison.round(3))

```

```python
# COMPARISON ANALYSIS:

# 01. CORRELATION - Measures linear relationships only
#       Max correlation: 0.30
#       Most correlations are weak (< 0.3)

# 02. PHIK - Measures any type of relationship (including non-linear)
#       Max phik : 0.42

#   This means variables may have non linear relationships with Salary
#   that correlation alone cannot capture!
```
